```python
import sklearn

# Print the installed version number to confirm success
print("Scikit-Learn version:", sklearn.__version__)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

# 1. Synthesize a complex, noisy classification dataset
X_fake, y_fake = make_classification(n_samples=1000, n_features=10, n_informative=5, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_fake, y_fake, test_size=0.2, random_state=42)

# 2. Train a single Decision Tree
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

# 3. Train a Random Forest with 100 parallel estimators (trees)
forest = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
forest.fit(X_train, y_train)

print(f"Single Tree Test Accuracy: {tree.score(X_test, y_test)*100:.2f}%")
print(f"Random Forest Test Accuracy: {forest.score(X_test, y_test)*100:.2f}%")

# 4. Extract Feature Importance
# Random Forests track exactly which variables contributed the most to reducing impurity
importances = forest.feature_importances_
for i, imp in enumerate(importances):
    print(f"Feature {i} Importance Score: {imp:.4f}")
```
