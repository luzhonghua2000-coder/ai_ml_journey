```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier


# 1. Create a raw mockup customer table with mixed structural types
raw_data = pd.DataFrame({
    'Age': [25, 47, 22, 35, 52, 30, 41, 28],
    'Income': [50000, 120000, 45000, 70000, 150000, 62000, 95000, 58000],
    'Membership': ['Silver', 'Gold', 'Silver', 'Bronze', 'Gold', 'Bronze', 'Silver', 'Bronze'],
    'BoughtPremium': [0, 1, 0, 0, 1, 0, 1, 0] # Label to predict
})

X = raw_data.drop(columns='BoughtPremium')
y = raw_data['BoughtPremium']

# 2. Segment feature column paths by processing needs
numeric_features = ['Age', 'Income']
categorical_features = ['Membership']

# 3. Create independent preprocessing sub-assemblies
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# 4. Merge sub-assemblies into a single structural transformer layout
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# 5. Pack the preprocessor and model together into a clean, unified Pipeline
pipeline_model = Pipeline(steps=[
    ('data_cleaning', preprocessor),
    ('classifier_engine', RandomForestClassifier(n_estimators=10, random_state=42))
])

# Train the whole pipeline in one step! Text handling and training happen automatically.
pipeline_model.fit(X, y)
predictions = pipeline_model.predict(X)

# 6. Evaluation Reports
print("\n=== Professional Classification Report ===")
print(classification_report(y, predictions))
```
