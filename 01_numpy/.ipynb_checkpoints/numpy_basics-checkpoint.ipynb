```python
# Generate a specific 2D matrix for practice
matrix = np.array([[10, 20, 30], 
                   [40, 50, 60], 
                   [70, 80, 90]])
```

```python
print(matrix)
```

```python
# 2. Slice out the first two rows and the last two columns
# Syntax is [start_row:end_row, start_col:end_col] (remember, end index is exclusive)
sub_matrix = matrix[0:2, 1:3]
print("Extracted Sub-matrix:\n", sub_matrix)
```

```python
# 3. Pull out the entire 3rd row
third_row = matrix[2, :]
print (third_row)
```

```python
# Generate an array of integers from 1 to 10
data = np.array([1, 8, 3, 12, 5, 7, 2, 14, 9, 4])
```

```python
# Step 1: Create the boolean mask condition
mask = data > 5
print("The Boolean Mask Array:\n", mask)
```

```python
# Step 2: Pass the mask back into the data array to filter out numbers instantly
filtered_data = data[mask]
print("\nFiltered Results (Only numbers > 5):\n", filtered_data)
```

```python
# Advanced: Replace specific numbers matching a condition
# Change any number less than 5 to a zero placeholder instantly
data[data < 5] = 0
print("\nData after in-place replacement:\n", data)
```

##:Task 2.1 (Scalar Stretch): Create a (3, 3) matrix containing numbers from 1 to 9. Multiply the entire matrix by 10 in a single line of code.

```python
import numpy as np
matrix = np.arange(1,10,1).reshape(3,3)
print ("Before:\n", matrix)
scalar =10
matrix = matrix * scalar
print ("After:\n", matrix)
```

## Task 2.2 (Matrix Addition): Create two separate (2, 4) matrices filled with random integers. Add them together so each overlapping cell updates its sum.

```python
import numpy as np
# 1. Create the first (2, 4) matrix filled with random integers between 1 and 10
# (low=1, high=11 means it includes numbers from 1 to 10. size=(2,4) sets the shape)
matrix_a = np.random.randint(low=1, high=11, size=(2, 4))

# 2. Create the second (2, 4) matrix filled with random integers between 1 and 10
matrix_b = np.random.randint(low=1, high=11, size=(2, 4))

# 3. Add them together (NumPy automatically pairs up overlapping cells)
matrix_sum = matrix_a + matrix_b

# 4. Print the results to see the math in action
print("Matrix A:\n", matrix_a)
print("\nMatrix B:\n", matrix_b)
print("\nResulting Sum Matrix (A + B):\n", matrix_sum)
```

##: Task 2.3 (Broadcasting Challenge): * Create a 2D matrix of shape (4, 3) filled with ones.

Create a 1D array of shape (3,) containing the numbers [10, 20, 30].

Add the 1D array to the 2D matrix. Observe how NumPy automatically "stretches" the row across all 4 rows of the matrix.

```python
import numpy as np
matrix_a =np.ones((4,3))
matrix_b =np.array([10, 20, 30]).reshape(3,)
matrix_c = matrix_a + matrix_b

print("Matrix A:\n", matrix_a)
print("\nMatrix B:\n", matrix_b)
print("\nResulting Sum Matrix (A + B):\n", matrix_c)
```

##: ask 3.1 (Basic Sieve): Create a 1D array containing the following random numbers: [12, 4, 7, 19, 2, 18, 5, 14, 8, 1].

Write a conditional statement to extract only the numbers that are strictly greater than 10.

```python
import numpy as np
data = np.array([12, 4, 7, 19, 2, 18, 5, 14, 8, 1])
selecteddata = data[data > 10]
print("Before:\n", data)
print("\nAfter:\n", selecteddata )
```

##: Task 3.2 (Even Number Extraction): Using the same array from Task 3.1, use the modulo operator (% 2 == 0) inside a boolean mask to filter and return only the even numbers.

```python
import numpy as np
data = np.array([12, 4, 7, 19, 2, 18, 5, 14, 8, 1])
selecteddata = data[data % 2 == 0]
print("Before:\n", data)
print("\nAfter:\n", selecteddata )
```

##: Task 3.3 (Data Clean-up / Boundary Cap): Create a (4, 4) matrix of random numbers between -5 and 5. In a single line of code, find every single negative number and replace it instantly with 0. (This mimics a critical deep learning function called a ReLU activation layer!).

```python
import numpy as np
matrix_a = np.random.randint(low=-5, high=5, size=(4, 4))

print("Before\n", matrix_a)

matrix_a [matrix_a < 0] = 0

print("After\n", matrix_a)
```

##: The Performance Verification Challenge

```python
import time
import numpy as np

# Create a large array of 1 million numbers
massive_array = np.arange(1_000_000)

# Method A: Traditional Python Loop
start_time = time.time()
loop_result = []
for x in massive_array:
    loop_result.append(x * 2)
print(f"Python Loop Time: {time.time() - start_time:.5f} seconds")

# Method B: NumPy Vectorization
start_time = time.time()
vector_result = massive_array * 2
print(f"NumPy Vectorized Time: {time.time() - start_time:.5f} seconds")
```
