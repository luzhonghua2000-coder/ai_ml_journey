```python
# Create two list-based matrices that are mathematically compatible: (2x3) * (3x2) -> Result should be (2x2)
A = [[1, 2, 3],
     [4, 5, 6]]

B = [[7, 8],
     [9, 1],
     [2, 3]]

# Initialize a 2x2 matrix with zeros to store the answer
result = [[0, 0],
          [0, 0]]

# Explicit matrix multiplication using nested loops
for i in range(len(A)):          # Loop through rows of A
    for j in range(len(B[0])):      # Loop through columns of B
        for k in range(len(B)):      # Loop through rows of B / columns of A
            result[i][j] += A[i][k] * B[k][j]

print("Loop Multiplication Result:\n", result)
```
