```python
# Create two list-based matrices that are mathematically compatible: (2x3) * (3x2) -> Result should be (2x2)
A = [[1, 2, 3],
     [4, 5, 6]]

B = [[7, 8],
     [9, 1],
     [2, 3]]

# Initialize a 2x2 matrix with zeros to store the answer
result = [[0, 0],
          [0, 0]]

# Explicit matrix multiplication using nested loops
for i in range(len(A)):          # Loop through rows of A
    for j in range(len(B[0])):      # Loop through columns of B
        for k in range(len(B)):      # Loop through rows of B / columns of A
            result[i][j] += A[i][k] * B[k][j]

print("Loop Multiplication Result:\n", result)
```

```python
# Now, see how NumPy does this in a single vectorized C-optimized line
import numpy as np
A_np = np.array(A)
B_np = np.array(B)

# The @ symbol or np.dot() is the vectorized matrix multiplication operator
numpy_result = A_np @ B_np
print("\nNumPy Vectorized Result:\n", numpy_result)
```

```python
import numpy as np

class LinearRegressionFromScratch:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        n_samples, n_features = X.shape
        # Initialize parameters to zeros
        self.weights = np.zeros(n_features)
        self.bias = 0.0
        
        # Optimization Loop (Gradient Descent)
        for _ in range(self.epochs):
            # 1. Forward Pass: Compute predictions
            y_predicted = np.dot(X, self.weights) + self.bias
            
            # 2. Compute Gradients (Partial Derivatives)
            dw = (2 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (2 / n_samples) * np.sum(y_predicted - y)
            
            # 3. Update Parameters (Take a step down the gradient)
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# --- Test Our Code With Toy Data ---
X_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([3, 5, 7, 9, 11]) # Rule: y = 2x + 1

model = LinearRegressionFromScratch(learning_rate=0.01, epochs=500)
model.fit(X_train, y_train)

print("Trained Weight (Expected approx 2):", model.weights[0])
print("Trained Bias (Expected approx 1):", model.bias)
```
