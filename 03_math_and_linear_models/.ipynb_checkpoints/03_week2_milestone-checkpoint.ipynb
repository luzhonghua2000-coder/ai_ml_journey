```python
import numpy as np
np.random.seed(42)
StudyHours = np.random.uniform(1.0, 10.0, (100, 1))
# If student studies > 5.5 hours, they pass (1), otherwise fail (0) with some added random noise
PassedExam = (StudyHours + np.random.normal(0, 1.0, (100, 1)) > 5.5).astype(int).flatten()
```

```python
class LogisticRegressionFromScratch:
    def __init__(self, learning_rate=0.05, epochs=1000):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        
    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
        
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0.0
        
        for _ in range(self.epochs):
            # 1. Linear combination wrapped in Sigmoid Activation
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self._sigmoid(linear_model)
            
            # 2. Compute Gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)
            
            # 3. Update weights
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
    def predict_proba(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return self._sigmoid(linear_model)
        
    def predict(self, X):
        # Convert continuous probabilities into explicit binary 0 or 1 classifications
        return [1 if p >= 0.5 else 0 for p in self.predict_proba(X)]
```

```python
# =====================================================================
# 3. TRAIN AND EVALUATE YOUR ENGINE
# =====================================================================
# Initialize our scratch model
model = LogisticRegressionFromScratch(learning_rate=0.1, epochs=3000)

# Train the parameters on our study data
model.fit(StudyHours, PassedExam)

# Generate final binary predictions across the dataset
predictions = model.predict(StudyHours)

# Compute classification accuracy (%)
accuracy = np.mean(predictions == PassedExam) * 100

print("\n=== Model Training Results ===")
print(f"Trained Weight (Slope Parameter): {model.weights[0]:.4f}")
print(f"Trained Bias (Y-Intercept): {model.bias:.4f}")
print(f"Final Model Classification Accuracy: {accuracy:.2f}%")

# Test custom edge-cases to observe the boundary
test_cases = np.array([[2.0], [5.5], [9.0]])
probabilities = model.predict_proba(test_cases)
print("\n=== Custom Predictions Matrix ===")
for hours, prob in zip(test_cases.flatten(), probabilities):
    status = "Pass" if prob >= 0.5 else "Fail"
    print(f"Studying {hours} hours -> {prob*100:.1f}% probability of passing -> Prediction: {status}")
```

```python
import numpy as np

# =====================================================================
# 1. PASTE YOUR LOGISTIC REGRESSION CLASS HERE
# =====================================================================
class LogisticRegressionFromScratch:
    def __init__(self, learning_rate=0.05, epochs=2000):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        
    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
        
    # Upgraded section inside your LinearRegressionFromScratch Class:
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0.0
    
        # Track history over time
        print("Beginning Training Pipeline...")
    
        for epoch in range(self.epochs):
            # Forward pass (Predictions Matrix)
            y_predicted = np.dot(X, self.weights) + self.bias
        
            # 1. SOLUTION: Calculate Mean Squared Error (MSE) Loss
            # Formula: Mean of (Errors Squared)
            mse_loss = np.mean((y_predicted - y) ** 2)
        
            # 2. SOLUTION: Print the tracking telemetry every 100 turns
            if epoch % 100 == 0:
                print(f"Epoch {epoch:04d} -> Current Mean Squared Error Loss: {mse_loss:.4f}")
        
            # Core Gradient Calculation and descent steps remain unchanged
            dw = (2 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (2 / n_samples) * np.sum(y_predicted - y)
        
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
    def predict_proba(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return self._sigmoid(linear_model)
        
    def predict(self, X):
        return np.array([1 if p >= 0.5 else 0 for p in self.predict_proba(X)])

# =====================================================================
# 2. GENERATE MOCK STUDY DATA (THE CHALLENGE SCENARIO)
# =====================================================================
np.random.seed(42)
# Generate 100 random student study hours between 1 and 10 hours
StudyHours = np.random.uniform(1.0, 10.0, (100, 1))

# Ground Truth Rule: If hours + noise > 5.5, they pass (1), else fail (0)
PassedExam = (StudyHours + np.random.normal(0, 1.0, (100, 1)) > 5.5).astype(int).flatten()

print(f"Dataset generated: {len(StudyHours)} students tracked.")
print(f"First 5 student study hours:\n{StudyHours[:5].flatten()}")
print(f"First 5 student exam outcomes (0=Fail, 1=Pass): {PassedExam[:5]}")

# =====================================================================
# 3. TRAIN AND EVALUATE YOUR ENGINE
# =====================================================================
# Initialize our scratch model
model = LogisticRegressionFromScratch(learning_rate=0.01, epochs=3000)

# Train the parameters on our study data
model.fit(StudyHours, PassedExam)

# Generate final binary predictions across the dataset
predictions = model.predict(StudyHours)

# Compute classification accuracy (%)
accuracy = np.mean(predictions == PassedExam) * 100

print("\n=== Model Training Results ===")
print(f"Trained Weight (Slope Parameter): {model.weights[0]:.4f}")
print(f"Trained Bias (Y-Intercept): {model.bias:.4f}")
print(f"Final Model Classification Accuracy: {accuracy:.2f}%")

# Test custom edge-cases to observe the boundary
test_cases = np.array([[2.0], [5.5], [9.0]])
probabilities = model.predict_proba(test_cases)
print("\n=== Custom Predictions Matrix ===")
for hours, prob in zip(test_cases.flatten(), probabilities):
    status = "Pass" if prob >= 0.5 else "Fail"
    print(f"Studying {hours} hours -> {prob*100:.1f}% probability of passing -> Prediction: {status}")
```
