```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic clustering dataset
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

class KMeansFromScratch:
    def __init__(self, k=3, max_iters=100, tol=1e-4):
        self.k = k
        self.max_iters = max_iters
        self.tol = tol
        self.centroids = None
        
    def fit(self, X):
        n_samples, n_features = X.shape
        
        # 1. Initialization: Choose random data points as starting centroids
        random_idx = np.random.choice(n_samples, self.k, replace=False)
        self.centroids = X[random_idx].copy()
        
        for i in range(self.max_iters):
            # 2. Assignment Step: Calculate Euclidean distances to centroids
            # Using broadcasting to compute distances efficiently
            # Shape of X: (n_samples, n_features)
            # Shape of centroids: (k, n_features)
            distances = np.sqrt(((X[:, np.newaxis, :] - self.centroids) ** 2).sum(axis=2))
            labels = np.argmin(distances, axis=1)
            
            # 3. Update Step: Calculate new centroids
            new_centroids = np.zeros_like(self.centroids)
            for cluster_idx in range(self.k):
                points_in_cluster = X[labels == cluster_idx]
                if len(points_in_cluster) > 0:
                    new_centroids[cluster_idx] = points_in_cluster.mean(axis=0)
                else:
                    # Handle empty cluster by keeping previous centroid
                    new_centroids[cluster_idx] = self.centroids[cluster_idx]
            
            # Check for convergence (if centroids stop shifting)
            centroid_shift = np.sum((self.centroids - new_centroids) ** 2)
            self.centroids = new_centroids.copy()
            
            if centroid_shift < self.tol:
                print(f"Converged early at iteration {i}")
                break
                
        return labels
    
    def predict(self, X):
        distances = np.sqrt(((X[:, np.newaxis, :] - self.centroids) ** 2).sum(axis=2))
        return np.argmin(distances, axis=1)

# Fit custom model
km_scratch = KMeansFromScratch(k=4, max_iters=50)
labels_scratch = km_scratch.fit(X)

# Fit Scikit-Learn Model for comparison
from sklearn.cluster import KMeans
km_sklearn = KMeans(n_clusters=4, init='random', n_init=1, max_iter=50, random_state=42)
labels_sklearn = km_sklearn.fit_predict(X)

# Plot Comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.scatter(X[:, 0], X[:, 1], c=labels_scratch, s=40, cmap='viridis')
ax1.scatter(km_scratch.centroids[:, 0], km_scratch.centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
ax1.set_title("K-Means Scratch Implementation")
ax1.legend()

ax2.scatter(X[:, 0], X[:, 1], c=labels_sklearn, s=40, cmap='viridis')
ax2.scatter(km_sklearn.cluster_centers_[:, 0], km_sklearn.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Centroids')
ax2.set_title("Scikit-Learn KMeans Implementation")
ax2.legend()

plt.tight_layout()
plt.show()
```
