### Arrays, Indexing, and Slicing

```python
import numpy as np
```

```python
# Create a 1D array of 5 zeros
array_1d = np.zeros(5)
print("1D Array:\n", array_1d)
print("Shape:", array_1d.shape)
```

```python
# Create a 2D array (3 rows, 4 columns) of all ones
array_2d = np.ones((3, 4))
print("\n2D Array:\n", array_2d)
print("Shape:", array_2d.shape)
```

```python
# Create a 3D array (2 matrices, each 3 rows by 3 columns) filled with random normal numbers
array_3d = np.random.randn(2, 3, 3)
print("\n3D Array:\n", array_3d)
print("Shape:", array_3d.shape)
```

```python
# Generate a specific 2D matrix for practice
matrix = np.array([[10, 20, 30], 
                   [40, 50, 60], 
                   [70, 80, 90]])
```

```python
print(matrix)
```

```python
# 2. Slice out the first two rows and the last two columns
# Syntax is [start_row:end_row, start_col:end_col] (remember, end index is exclusive)
sub_matrix = matrix[0:2, 1:3]
print("Extracted Sub-matrix:\n", sub_matrix)
```

```python
# 3. Pull out the entire 3rd row
third_row = matrix[2, :]
print (third_row)
```

```python
# 2. Add two matrices of identical shapes together instantly
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
matrix_sum = matrix_a + matrix_b
print("Element-wise sum:\n", matrix_sum)
```

```python
# Generate an array of integers from 1 to 10
data = np.array([1, 8, 3, 12, 5, 7, 2, 14, 9, 4])
```

```python
# Step 1: Create the boolean mask condition
mask = data > 5
print("The Boolean Mask Array:\n", mask)
```

```python
# Step 2: Pass the mask back into the data array to filter out numbers instantly
filtered_data = data[mask]
print("\nFiltered Results (Only numbers > 5):\n", filtered_data)
```

```python
# Advanced: Replace specific numbers matching a condition
# Change any number less than 5 to a zero placeholder instantly
data[data < 5] = 0
print("\nData after in-place replacement:\n", data)
```
